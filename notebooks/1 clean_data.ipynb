```python
import pandas as pd
```

```python
def describe_df(df):
    descriptive_summary = df.describe(include="all").T
    descriptive_summary["n_unique"] = df.nunique()
    descriptive_summary["null_count"] = df.isnull().sum()
    descriptive_summary["skewness"] = df.skew(numeric_only=True)
    descriptive_summary["kurtosis"] = df.kurtosis(numeric_only=True)
    descriptive_summary["type"] = df.dtypes
    
    try:
        descriptive_summary = descriptive_summary.drop(columns=["unique", "top", "freq"])
    except:
        pass

    return descriptive_summary
```

```python
ventas_cdmx_df = pd.read_excel('../data/_raw/ventas_CDMX_aspel_export.xlsx', skiprows=3)
ventas_cdmx_df
```

```python
# FECHA
ventas_cdmx_df['ERROR_FECHA'] = ventas_cdmx_df['FECHA'].apply(
    lambda x: x if pd.to_datetime(x, errors='coerce') is pd.NaT else None
)

errores_fecha_unicos = ventas_cdmx_df['ERROR_FECHA'].dropna().unique()

print("Valores únicos que no pudieron convertirse:")
print(errores_fecha_unicos)

ventas_cdmx_df['FECHA'] = pd.to_datetime(ventas_cdmx_df['FECHA'], errors='coerce')
```

```python
# PRECIO_UNIT
ventas_cdmx_df['ERROR_PRECIO_UNIT'] = ventas_cdmx_df['PRECIO_UNIT'].apply(
    lambda x: x if pd.isna(pd.to_numeric(x, errors='coerce')) else None
)

errores_precio_unicos = ventas_cdmx_df['ERROR_PRECIO_UNIT'].dropna().unique()

print("Valores únicos que no pudieron convertirse:")
print(errores_precio_unicos)

ventas_cdmx_df['PRECIO_UNIT'] = pd.to_numeric(ventas_cdmx_df['PRECIO_UNIT'], errors='coerce')
```

```python
# SUBTOTAL
ventas_cdmx_df['ERROR_SUBTOTAL'] = ventas_cdmx_df['SUBTOTAL'].apply(
    lambda x: x if pd.isna(pd.to_numeric(x, errors='coerce')) else None
)

errores_subtotal_unicos = ventas_cdmx_df['ERROR_SUBTOTAL'].dropna().unique()

print("Valores únicos que no pudieron convertirse:")
print(errores_subtotal_unicos)

ventas_cdmx_df['SUBTOTAL'] = pd.to_numeric(ventas_cdmx_df['SUBTOTAL'], errors='coerce')
```

```python
# IVA
ventas_cdmx_df['ERROR_IVA'] = ventas_cdmx_df['IVA'].apply(
    lambda x: x if pd.isna(pd.to_numeric(x, errors='coerce')) else None
)

errores_iva_unicos = ventas_cdmx_df['ERROR_IVA'].dropna().unique()

print("Valores únicos que no pudieron convertirse:")
print(errores_iva_unicos)

ventas_cdmx_df['IVA'] = pd.to_numeric(ventas_cdmx_df['IVA'], errors='coerce')
```

```python
# TOTAL
ventas_cdmx_df['ERROR_TOTAL'] = ventas_cdmx_df['TOTAL'].apply(
    lambda x: x if pd.isna(pd.to_numeric(x, errors='coerce')) else None
)

errores_total_unicos = ventas_cdmx_df['ERROR_TOTAL'].dropna().unique()

print("Valores únicos que no pudieron convertirse:")
print(errores_total_unicos)

ventas_cdmx_df['TOTAL'] = pd.to_numeric(ventas_cdmx_df['TOTAL'], errors='coerce')
```

```python
# MY TOTAL
ventas_cdmx_df['my_total'] = ventas_cdmx_df['CANTIDAD'] * ventas_cdmx_df['PRECIO_UNIT'] * 1.16
```

```python
ventas_cdmx_df = ventas_cdmx_df.drop(columns=['ERROR_FECHA', 'ERROR_PRECIO_UNIT', 'ERROR_SUBTOTAL', 'ERROR_IVA', 'ERROR_TOTAL'])
descriptive_summary = describe_df(ventas_cdmx_df)
descriptive_summary
```
