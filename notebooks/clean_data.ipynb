```python
import pandas as pd
```

```python
def describe_df(df):
    descriptive_summary = df.describe(include="all").T
    descriptive_summary["n_unique"] = df.nunique()
    descriptive_summary["null_count"] = df.isnull().sum()
    descriptive_summary["skewness"] = df.skew(numeric_only=True)
    descriptive_summary["kurtosis"] = df.kurtosis(numeric_only=True)
    descriptive_summary["type"] = df.dtypes
    
    try:
        descriptive_summary = descriptive_summary.drop(columns=["unique", "top", "freq"])
    except:
        pass

    return descriptive_summary
```

```python
ventas_gdl_2023_df = pd.read_excel('../data/_raw/ventas_GDL_lupita_manual.xlsx', sheet_name='Ventas 2023')
ventas_gdl_2024_df = pd.read_excel('../data/_raw/ventas_GDL_lupita_manual.xlsx', sheet_name='Ventas 2024')

ventas_gdl_df = pd.concat([ventas_gdl_2023_df, ventas_gdl_2024_df], ignore_index=True)
new_ventas_gdl_df = ventas_gdl_df.copy()
ventas_gdl_df
```

```python
# FECHA
new_ventas_gdl_df = new_ventas_gdl_df.rename(columns={'Fecha': 'FECHA'})

new_ventas_gdl_df['ERROR_FECHA'] = new_ventas_gdl_df['FECHA'].apply(
    lambda x: x if pd.to_datetime(x, errors='coerce', format='mixed', dayfirst=True) is pd.NaT else None
)

errores_fecha_unicos = new_ventas_gdl_df['ERROR_FECHA'].dropna().unique()

print("Valores únicos que no pudieron convertirse:")
print(errores_fecha_unicos)

new_ventas_gdl_df['FECHA'] = pd.to_datetime(new_ventas_gdl_df['FECHA'], errors='coerce', format='mixed', dayfirst=True)

new_ventas_gdl_df = new_ventas_gdl_df.dropna(subset='FECHA')
```

```python
# PRODUCTO
new_ventas_gdl_df = new_ventas_gdl_df.rename(columns={'Producto': 'PRODUCTO'})
productos_unicos = new_ventas_gdl_df['PRODUCTO'].dropna().unique()
print("Productos unicos:")
print(productos_unicos)
```

```python
# CANTIDAD
new_ventas_gdl_df = new_ventas_gdl_df.rename(columns={'Ctdad': 'CANTIDAD'})
new_ventas_gdl_df['ERROR_CANTIDAD'] = new_ventas_gdl_df['CANTIDAD'].apply(
    lambda x: x if pd.isna(pd.to_numeric(x, errors='coerce')) else None
)

errores_cantidad_unicos = new_ventas_gdl_df['ERROR_CANTIDAD'].dropna().unique()

print("Valores únicos que no pudieron convertirse:")
print(errores_cantidad_unicos)

new_ventas_gdl_df['CANTIDAD'] = pd.to_numeric(new_ventas_gdl_df['CANTIDAD'], errors='coerce')
```

```python
# PRECIO_UNIT
new_ventas_gdl_df = new_ventas_gdl_df.rename(columns={'P. Unit': 'PRECIO_UNIT'})

new_ventas_gdl_df['PRECIO_UNIT'] = (
    new_ventas_gdl_df['PRECIO_UNIT']
    .astype(str)
    .str.replace(r'[\$,]', '', regex=True)
    .str.strip()
)

new_ventas_gdl_df['ERROR_PRECIO_UNIT'] = new_ventas_gdl_df['PRECIO_UNIT'].apply(
    lambda x: x if pd.isna(pd.to_numeric(x, errors='coerce')) else None
)

errores_precio_unicos = new_ventas_gdl_df['ERROR_PRECIO_UNIT'].dropna().unique()

print("Valores únicos que no pudieron convertirse:")
print(errores_precio_unicos)

new_ventas_gdl_df['PRECIO_UNIT'] = pd.to_numeric(new_ventas_gdl_df['PRECIO_UNIT'], errors='coerce')
```

```python
# TOTAL
new_ventas_gdl_df = new_ventas_gdl_df.rename(columns={'Total c/IVA': 'TOTAL'})

new_ventas_gdl_df['TOTAL'] = (
    new_ventas_gdl_df['TOTAL']
    .astype(str)
    .str.replace(r'[\$,]', '', regex=True)
    .str.strip()
)

new_ventas_gdl_df['ERROR_TOTAL'] = new_ventas_gdl_df['TOTAL'].apply(
    lambda x: x if pd.isna(pd.to_numeric(x, errors='coerce')) else None
)

errores_precio_unicos = new_ventas_gdl_df['ERROR_TOTAL'].dropna().unique()

print("Valores únicos que no pudieron convertirse:")
print(errores_precio_unicos)

new_ventas_gdl_df['TOTAL'] = pd.to_numeric(new_ventas_gdl_df['TOTAL'], errors='coerce')
```

```python
# FORMA_PAGO
new_ventas_gdl_df = new_ventas_gdl_df.rename(columns={'Forma de pago': 'FORMA_PAGO'})

new_ventas_gdl_df['FORMA_PAGO'] = (
    new_ventas_gdl_df['FORMA_PAGO']
    .str.strip()
    .str.lower()
)

map_fp = {
    'efectivo': 'Efectivo',
    'tarjeta': 'Tarjeta',
    'credito': 'Crédito 30 días',
    'transferencia': 'Transferencia',
    'transf': 'Transferencia',
    'cheque': 'Cheque'
}

new_ventas_gdl_df['FORMA_PAGO'] = new_ventas_gdl_df['FORMA_PAGO'].map(map_fp)

fp_unicos = new_ventas_gdl_df['FORMA_PAGO'].dropna().unique()
print("fp unicos:")
print(fp_unicos)
```

```python
# CLIENTE
new_ventas_gdl_df = new_ventas_gdl_df.rename(columns={'Cliente': 'CLIENTE'})
```

```python
# Vendedor
new_ventas_gdl_df = new_ventas_gdl_df.rename(columns={'Vendedor': 'VENDEDOR'})

new_ventas_gdl_df['VENDEDOR'] = (
    new_ventas_gdl_df['VENDEDOR']
    .str.strip()
    .str.lower()
)

map_v = {
    'rod': 'Rodrigo',
    'rodrigo v.': 'Rodrigo',
    'lupita': 'Lupita',
    'lupita m.': 'Lupita'
}

new_ventas_gdl_df['VENDEDOR'] = new_ventas_gdl_df['VENDEDOR'].map(map_v)

v_unicos = new_ventas_gdl_df['VENDEDOR'].dropna().unique()
print("VENDEDOR unicos:")
print(v_unicos)
```

```python
new_ventas_gdl_df = new_ventas_gdl_df.drop(columns=['ERROR_FECHA', 'ERROR_CANTIDAD', 'Unidad', 'ERROR_PRECIO_UNIT', 'ERROR_TOTAL', 'Notas / Obs'])
descriptive_summary = describe_df(new_ventas_gdl_df)
descriptive_summary
```

```python
gastos_operativos_df = pd.read_excel('../data/_raw/gastos_operativos_dueño.xlsx')
gastos_operativos_df
```

```python
# FECHA y Periodo
gastos_operativos_df = gastos_operativos_df.rename(columns={'fecha': 'FECHA', 'periodo': 'PERIODO'})

gastos_operativos_df['FECHA'] = pd.to_datetime(gastos_operativos_df['FECHA'], errors='coerce', format='mixed', dayfirst=True)

gastos_operativos_df['ERROR_PERIODO'] = gastos_operativos_df['PERIODO'].apply(
    lambda x: x if pd.to_datetime(x, errors='coerce', format='mixed', dayfirst=True) is pd.NaT else None
)

errores_periodo_unicos = gastos_operativos_df['ERROR_PERIODO'].dropna().unique()

print("Valores únicos que no pudieron convertirse:")
print(errores_periodo_unicos)
```

```python
format_especifico = pd.to_datetime(gastos_operativos_df['PERIODO'], format='%b-%y', errors='coerce')
format_flexible = pd.to_datetime(gastos_operativos_df['PERIODO'], errors='coerce', dayfirst=True, format='mixed')

gastos_operativos_df['PERIODO'] = format_especifico.fillna(format_flexible)
```

```python
# CONCEPTO y MONTO
gastos_operativos_df = gastos_operativos_df.rename(columns={'concepto': 'CONCEPTO', 'monto': 'MONTO'})
concepto_unicos = gastos_operativos_df['CONCEPTO'].dropna().unique()
print("concepto unicos:")
print(concepto_unicos)
```

```python
# COMPROBANTE y OBSERVACIONES
gastos_operativos_df = gastos_operativos_df.rename(columns={'comprobante': 'COMPROBANTE', 'observaciones': 'OBSERVACIONES'})
comprobante_unicos = gastos_operativos_df['COMPROBANTE'].dropna().unique()
print("comprobante unicos:")
print(comprobante_unicos)
```

```python
# FORMA_PAGO
gastos_operativos_df = gastos_operativos_df.rename(columns={'pagado_por': 'FORMA_PAGO'})
pagado_por_unicos = gastos_operativos_df['FORMA_PAGO'].dropna().unique()
print("FORMA_PAGO unicos:")
print(pagado_por_unicos)
```

```python
gastos_operativos_df = gastos_operativos_df.drop(columns=['mes_contable', 'ERROR_PERIODO'])
descriptive_summary = describe_df(gastos_operativos_df)
descriptive_summary
```
